##Everything's Random

Now that you know a little about how to create & manipulate matrices in NumPy, let's see how we'd generate _**totally random matrices**_ in NumPy.

NumPy has a really cool [**random**](https://docs.scipy.org/doc/numpy/reference/random/generator.html) module. Using ```np.random``` we can create random integer arrays and random floating point arrays. (_Again only basics here, please check out the **resources** for details._)

<br>

#### There are 3 main random functions : 
1. **randint(low,high,size):** This gives random interger arrays.

2. **randn(size):** This gives random floating point numbers based on Normal Distribution. In normal distributions as discussed in Part-I, mean=0 and variance=1.The size here represents the dimensions of the returned array. If nothing is passed we will get a single floating-point number as output.

3. **rand(size):**  This gives random floating point numbers based on the uniform distribution valued between [0,1). (Note the open and closed interval).The size here represents the dimensions of the returned array. If nothing is passed we will get a single floating-point number as output.


There's a difference between `randn()` and `rand()`. 

The array created using `rand()`is filled with _**random samples from a uniform distribution over [0, 1]**_ 


whereas the array created using `randn()` is filled with _**random values from normal distribution**_.  

<br>

Generally, we use `randint()` when we need random values but **integers.** `randn()` on the other hand is used when we want **floating-point** random numbers that are both positive & negative. Finally, the `rand()` function (_unlike `randn()`_) is used when we want random floating-point numbers that are **only positive** & in the range [0,1).  

<br>

**Let's see some examples:**


## 1. ```randint()```

````
rand_int=np.random.randint(2,10)  # if the shape is not mentioned the output will just be a random integer in the given range
print(rand_int)

rand_int2=np.random.randint(1,100,(4,5)) #random numpy array of shape (4,5)
print(rand_int2)

rand_int3=np.random.randint(50,75,(2,2),dtype='int64')
print(rand_int3)
````

##2. ```randn()```

````
rand_n=np.random.randn() #outputs a single floating point number
print(rand_n)

rand_n2=np.random.randn(4,5) #outputs an 4x5 array filled with random floating point numbers 
print(rand_n2)
````

## 3. ```rand()```

````

rand=np.random.rand() #outputs single floating-point value .
print(rand)

rand2=np.random.rand(3,2) #outputs an 3x2 array filled with random floating point values.
print(rand2)
````

# 1. Normal Distribution(Gaussian Distribution) :

Recall what this distribution is: mean is **0** and variance is **1**.

We'll make the underlying equation first using **NumPy** & then use **MatPlotlib** to visualize that equation.


````
#Just setting up Matplotlib 

%matplotlib inline

import matplotlib.pyplot as plt

plt.style.use('seaborn') # pretty matplotlib plots
plt.rcParams['figure.figsize'] = (12, 8)
````

````
x=np.linspace(-5,5,1000) #linspace will generate 1000 values linearly spaced from -5 to 5 
mean=0 #greek letter mu
variance=1 #greek letter sigma

#Now I am going to write the scaring PDF we saw in Part -I but you need to focus on plot :)
y = (1 / (np.sqrt(2 * np.pi * np.power(variance, 2)))) * \
    (np.power(np.e, -(np.power((x - mean), 2) / (2 * np.power(variance, 2)))))
````

````
#On to the plotting 
'''
Observe how it's centered at 0 which shows mean/average is 0.
'''
plt.plot(x,y)
plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.show()
````

_But an even **easier** way_ is to use Numpy's in-built function [**normal**](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.normal.html).


We can also use **seaborn's** [**distplot**](https://pythonbasics.org/seaborn-distplot/) to show how the curve overlays the distribution, but you can also use `histogram` as we did earlier.

````
import seaborn as sns
x=np.random.normal(0,1,1000)
'''
#To use histogram's plot uncomment this part
plt.plot(x)
plt.hist(x,rwidth=0.3)
plt.show()
'''
#using Seaborn's distplot
sns.distplot(x)
plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
````

# **3. Bernoulli Distribution**




Remember that the **Bernoulli Distribution** is a case of a **Binomial Distribution** where we only conduct a _**single experiment**_.   It's als a _**discrete probability distribution**_ with `probability p` for value 1 & `probability q=1-p` for value 0. 

<br>

p can be for success, yes, true, or one.  
Similarly, q=1-p can be for failure, no, false, or zero.

**Dive Deeper into this with :** https://towardsdatascience.com/fun-with-the-binomial-distribution-96a5ecabf65b



````
bernoulli=np.random.binomial(50,0.3,1)
sns.distplot(bernoulli)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Bernoulli Distribution')
plt.show()
````

## **4. Binomial Distribution**

````
bernoulli=np.random.binomial(1000,0.7,100)
sns.distplot(bernoulli)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Binomial Distribution')
plt.show()
````

#**5.Poissons Distribution**


The Poisson distribution is the limit of the binomial distribution for large N.



````
x=np.random.poisson(5,1000)
sns.distplot(x)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Poissons Distribution')
plt.show()
````
